```python
########
######## This block generates individual plots for all terms / countries separated by journal.
########

# The search term(s)
for term in countries:

    # the data we want to plot
    data = analyze(INPUT_FILE, False, [term], None, None)

    years = list()
    journal_1 = list()
    journal_2 = list()
    journal_3 = list()
    journal_4 = list()
    journal_5 = list()

    for year in data:
        years.append(year)
        for journal in data[year]:

            if journal == "The Middle East Journal":
                journal_1.append( data[year][journal]['matches_percent'] )

            if journal == "Middle East Policy":
                journal_2.append( data[year][journal]['matches_percent'] )

            if journal == "Middle Eastern Studies":
                journal_3.append( data[year][journal]['matches_percent'] )

            if journal == "British Journal of Middle Eastern Studies":
                journal_4.append( data[year][journal]['matches_percent'] )

            if journal == "International Journal of Middle East Studies":
                journal_5.append( data[year][journal]['matches_percent'] )

    appdata = ColumnDataSource(data=dict(years=years, journal_1=journal_1, journal_2=journal_2, 
                                         journal_3=journal_3, journal_4=journal_4, journal_5=journal_5))

    # create a new plot with a title and axis labels
    p = figure(
        title=str(term), 
        x_axis_label='Year', 
        y_axis_label='% of abstracts containing "'+str(term)+'"', 
        plot_width=900)


    #output_notebook()
    output_file(term+".html")

    # add a line renderer with legend and line thickness
    p.circle('years', 'journal_1', legend="The Middle East Journal", source=appdata, line_color="orange", fill_alpha=0.2, size=10)
    p.line('years', 'journal_1', source=appdata, line_width=1, line_color="orange")

    p.circle('years', 'journal_2', legend="Middle East Policy", source=appdata, line_color="purple", fill_alpha=0.2, size=10)
    p.line('years', 'journal_2', source=appdata, line_width=1, line_color="purple")

    p.circle('years', 'journal_3', legend="Middle Eastern Studies", source=appdata, line_color="blue", fill_alpha=0.2, size=10)
    p.line('years', 'journal_3', source=appdata, line_width=1, line_color="blue")

    p.circle('years', 'journal_4', legend="British Journal of Middle Eastern Studies", source=appdata, line_color="red", fill_alpha=0.2, size=10)
    p.line('years', 'journal_4', source=appdata, line_width=1, line_color="red")

    p.circle('years', 'journal_5', legend="International Journal of Middle East Studies", source=appdata, line_color="green", fill_alpha=0.2, size=10)
    p.line('years', 'journal_5', source=appdata, line_width=1, line_color="green")

    # show the results
    show(p)
```
